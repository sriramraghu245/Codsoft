```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, recall_score , classification_report
```

```python
cc_data= pd.read_csv('/content/creditcard.csv')
```

```python
cc_data.head()
```

```python
legit=cc_data[cc_data.Class==0]
fraud=cc_data[cc_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
legit_sample=legit.sample(n=261)
```

```python
new_data=pd.concat([legit_sample, fraud],axis=0)
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X=new_data.drop(columns='Class', axis=1)
Y=new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split

# X is your feature data, and Y is your target variable
# Replace X and Y with your actual data

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)

# The 'test_size' parameter determines the size of the testing set (20% in this case).
# 'stratify=Y' ensures that the class distribution in the training and testing sets is similar.
# 'random_state=42' is a seed for the random number generator, ensuring reproducibility.

# You can now use X_train and Y_train for training your model
# and X_test and Y_test for evaluating its performance.
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data :{:.2f}%'.format(training_data_accuracy*100))
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data :{:.2f}%'.format(test_data_accuracy*100))
```

```python
print (classification_report(Y_test , model.predict(X_test)))
```
